```python
def make_df(cols, ind):
    """Быстро создаем объект DataFrame"""
    data = {c: [str(c) + str(i) for i in ind]
            for c in cols}
    return pd.DataFrame(data, ind)
```

```python
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),'data2': rng.randint(0, 10, 6)},columns = ['key', 'data1', 'data2'])
df
```

```python
# aggregate allows more, because he can compute other functions
df.groupby('key').aggregate(['min', 'median', 'max'])
```

```python
# also you can apply other functions for different columns!
df.groupby('key').aggregate({'data1': ['min', 'median'], 'data2': 'max'})
```

```python
# you can apply varied filters onto dataframe
def filter_func(x):
    return x['data2'].std() > 4 # we sift all values, when standart deviation is critical

df.groupby('key').aggregate(filter_func) # we see, that A not become
```

```python
df.groupby('key').filter(filter_func)
```

```python
# transform
# aggragate return abbreviated data, and transform can return all data
df.groupby('key').transform(lambda x: x - x.mean())
```

```python
# apply can applied onto results of grouping
def norm_by_data2(x):
    x['data1'] /= x['data2'].sum()
    return x

df.groupby('key').apply(norm_by_data2)
```

```python
nrows, ncols = 100000, 100
rng = np.random.RandomState(42)
df1, df2, df3, df4 = (pd.DataFrame(rng.rand(nrows, ncols)) for i in range(4))
```

```python
%timeit df1 + df2 + df3 + df4
```

```python
%timeit pd.eval('df1 + df2 + df3 + df4')
```

```python
df1, df2, df3, df4, df5 = (pd.DataFrame(rng.randint(0, 1000, (100, 3))) for i in range(5))
```

```python
#pd eval can ariphmetical operations
%timeit (-df1) * df2 / (df3 + df4) - df5
%timeit pd.eval('-df1 * df2 / (df3 + df4) - df5')
```

```python
%timeit (df1 < df2) & (df2 <= df3) & (df3 != df4)
%timeit pd.eval('(df1 < df2) & (df2 <= df3) & (df3 != df4)')
```

```python
df = pd.DataFrame(rng.rand(1000, 3), columns=['A', 'B', 'C'])
df.eval('D = (A + B) / C', inplace=True)

%timeit df[(df.A < 0.5) & (df.B < 0.5)]
%timeit pd.eval('df[(df.A < 0.5) & (df.B < 0.5)]')
%timeit df.query('A < 0.5 and B < 0.5')

# so, eval are not faster, then standart operations
```

```python
Cmean = df['C'].mean()
%timeit df[(df.A < Cmean) & (df.B < Cmean)]
%timeit df.query('A < @Cmean and B < @Cmean')
```

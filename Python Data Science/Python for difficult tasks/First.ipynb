```python
def sum_of_lists(N):
    total = 0
    for i in range(5):
        L = [j ^ (j>>i) for j in range(N)]
        total += sum(L)
    return total
```

```python
%prun sum_of_lists(100000)
```

```python
%%file mprun_demo.py
def sum_of_lists(N):
    total = 0
    for i in range(5):
        L = [j ^ (j >> i) for j in range(N)]
        total += sum(L)
        del L # Удалить ссылку на L
        return total
```

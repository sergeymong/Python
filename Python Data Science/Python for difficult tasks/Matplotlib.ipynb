```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

# you can use cut names
plt.plot(x, x + 4, linestyle='-')
plt.plot(x, x + 5, linestyle='--')
plt.plot(x, x + 6, linestyle='-.')
plt.plot(x, x + 7, linestyle=':');
```

```python
# easy way to change axies
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
# if you need mirrored output you can use this
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.5, -1.5);
```

```python
# plt.axis -- simplified syntax
plt.plot(x, np.sin(x))
plt.axis([10, 0, 1.5, -1.5]); # [xmin, xmax, ymin, ymax]
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight'); 
# you can use automatically settings
```

```python
# or
plt.plot(x, np.sin(x))
plt.axis('equal'); 
```

```python
# to set titles and names use this
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)');

# also you can change position this titles and labels
```

```python
# also you can add legend
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('functions')
plt.axis('equal')
plt.legend();
```

```python
# methods for object-oriented style
# plt.xlabel()ax.set_xlabel();
# plt.ylabel()ax.set_ylabel();
# plt.xlim()ax.set_xlim();
# plt.ylim()ax.set_ylim();
# plt.title()ax.set_title().

# in the object-oriented style simpler use ax.set for all this methods
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
      xlabel='x', ylabel='sin(x)',
      title = 'A simple piot');
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);

# all type of markers
```

```python
# plt.scatter specially was created for scatter plots
plt.scatter(x, y);
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar(); # this is optionally, we show colorbar
```

```python
from sklearn.datasets import load_iris
iris = load_iris()

features = iris.data.T # transpose data
features
```

```python
plt.scatter(features[0], features[1], alpha=0.2, s=features[3]*100, c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

```python
# in the big datasets plt.plot much faster than plt.scatter
```

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='.k'); # yerr and fmt set styles. fmt -- style of lines, 
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black', ecolor='lightgray', elinewidth=3, capsize=0); 
# capsize -- ends of errors, elinewidth -- width of errorline, ecolor -- color of error
```
